```python
import matplotlib.pyplot as plt

def draw_neural_network():
    fig, ax = plt.subplots(figsize=(8, 6))

    # Neurons
    input_neurons = 8
    hidden_neurons = 6
    output_neurons = 1

    # X
    x_input = 0.1
    x_hidden = 0.5
    x_output = 0.9

    # Y
    input_y = range(input_neurons)
    hidden_y = range(hidden_neurons)
    output_y = range(output_neurons)

    # Draw neurons
    for y in input_y:
        ax.scatter(x_input, y, s=800)

    for y in hidden_y:
        ax.scatter(x_hidden, y, s=800)

    for y in output_y:
        ax.scatter(x_output, y, s=800)

    # Draw connections
    for i in input_y:
        for h in hidden_y:
            ax.plot([x_input, x_hidden], [i, h])

    for h in hidden_y:
        for o in output_y:
            ax.plot([x_hidden, x_output], [h, o])

    # Labels
    ax.text(x_input, max(input_y)+1, "Input Layer\n(8 Features)", ha='center')
    ax.text(x_hidden, max(hidden_y)+1, "Hidden Layer\n(6 Neurons)", ha='center')
    ax.text(x_output, max(output_y)+1, "Output Layer\n(1 Neuron)", ha='center')

    ax.axis('off')
    plt.title("Neural Network for Pima Indians Diabetes Prediction")
    plt.show()

draw_neural_network()
```
