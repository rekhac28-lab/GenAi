```python
import matplotlib.pyplot as plt

def draw_nn():
    fig, ax = plt.subplots()

    input_neurons = 6
    hidden_neurons = 5
    output_neurons = 1

    x_input = 0.1
    x_hidden = 0.5
    x_output = 0.9

    for i in range(input_neurons):
        ax.scatter(x_input, i)

    for i in range(hidden_neurons):
        ax.scatter(x_hidden, i)

    for i in range(output_neurons):
        ax.scatter(x_output, i)

    for i in range(input_neurons):
        for j in range(hidden_neurons):
            ax.plot([x_input, x_hidden], [i, j])

    for i in range(hidden_neurons):
        ax.plot([x_hidden, x_output], [i, 0])

    ax.text(x_input, input_neurons, "Input Layer")
    ax.text(x_hidden, hidden_neurons, "Hidden Layer")
    ax.text(x_output, output_neurons, "Output Layer")

    ax.axis("off")
    plt.title("Neural Network for Fraud Detection")
    plt.show()

draw_nn()
```
